```python
def fibonacci():
    '''Infinite fibonacci series'''
    x, y = 0, 1
    while True:
        yield x
        x, y = y, x + y
```

```python
from __future__ import print_function
g = fibonacci()
for _ in range(34):
    print('{:,}'.format(next(g)), end=' ')
```

```python
import itertools
```

```python
for fib in itertools.islice(fibonacci(), 50):
    print('{:,}'.format(fib), end='; ')
```
